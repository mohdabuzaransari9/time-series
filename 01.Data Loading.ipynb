**Querying by time**

```python
series["1959"]  # series offer this functionality only by input like this it gives all the values of that year
```

```python
series["1959-1"] # series offer this functionality only by input like this it gives all the values of that month and it also applies on date too
```

```python
# on the other hand dataframe provides condition
data2[(data2['date']>'1959-01-01') & (data2["date"]<="1959-01-21")]

# in a square bracket
# separate condition in parenthesis
# combined with &
```

```python
data3[(data3.index > '1959-01-01') & (data3.index <= '1959-01-21')]
```
