## **decomposing**

we can create a time series data as a combination of four different parts.

Level -> average value in series

trend -> increasing or decreasing value in series

Seasonality -> repeating short term cycle in series

noise -> random variation in series








**two ways to combine these four parts to create time series data**

```python
import pandas as pd
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
miles_decomp_df = pd.read_csv('data/us-airlines-monthly-aircraft-miles-flown.csv',header=0,parse_dates=[0])
```

```python
miles_decomp_df.head()
```

```python
miles_decomp_df.index=miles_decomp_df["Month"]
```

```python
miles_decomp_df.head()
```

```python
result=seasonal_decompose(miles_decomp_df['MilesMM'],model="additive")
# provide two parameters -> series with values & model = additive or multiplicative
```

**observed** -> line chart of our original data series.


**trend** -> line chart give us the trend values after removing noise and seasonality

**seasonal** -> shows seasonality trend

**residual** -> remaining left after combining (noise and seasonality)

you have to look at y axis as well.



```python
result.plot() # check also y axis here
```

seasonality variation in our data is from -2000 unit to +2000 unit

```python
result2=seasonal_decompose(miles_decomp_df['MilesMM'],model="multiplicative")
# provide two parameters -> series with values & model = additive or multiplicative
```

```python
result2.plot() # check also y axis here
```

seasonality -> y axis varies from 0.8 to 1.2

because we are using multiplicative model here.

for additive seasonality, we were adding seasonality value.

That's why we were getting -2000 to 2000.

But here we are multiplying the seasonality.

So 0.8 means we are reducing the value of y(t)

and 1.2 means we are increasing the value of y(t) by 20 percent.

in additive model -> We using absolute value to add on
in multiplicative model  -> we are multiplying that valuewith our data.



**easy way to identify that in your data you have seaonality and trend yes or no is called decomposing**
